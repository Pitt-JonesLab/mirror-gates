```python
from qiskit import QuantumCircuit

N = 4
qc = QuantumCircuit(N)
for i in range(N - 1):
    qc.cx(i, N - 1)
qc.draw("latex", filename="base.png")
```

```python
qc = QuantumCircuit(4)
qc.cx(1, 2)
qc.swap(1, 2)
qc.swap(2, 3)
qc.cx(2, 1)
qc.cx(0, 1)
qc.draw("latex", filename="base.png")
```
